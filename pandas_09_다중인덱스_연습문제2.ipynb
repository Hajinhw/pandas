## 나주시 2004년~ 2020년 데이터

```python
naju1 = pd.read_csv('data/나주시_주택현황 및 보급률(2004~2020).csv', encoding='cp949', header=[0,1,2])
naju2 = pd.read_csv('data/나주시_주택현황 및 보급률_2021년.csv', encoding='cp949')
```

```python
naju1
```

```python
naju1_T = naju1.T
naju1_T = naju1_T.unstack(level=0)
naju1_T
```

```python
# naju1_T.drop(['행정구역(시군)별(1)','행정구역(시군)별(1)'], inplace=True)
# naju1_T.drop(columns=[(0,'행정구역(시군)별(1)')], inplace=True)
naju1 = naju1_T.droplevel(level=0, axis=1)
naju1
```

```python
index = ['일반가구수(A)(가구)', '주택보급률(B)/(A)*100(%)', '다세대주택(호)', '단독주택', '다가구주택(호)', 
         '비거주용건물내주택(호)', '주택수(B)(호)', '아파트(호)', '연립주택(호)']
```

```python
naju1.index = index
naju1
```

## 나주시 2021년 데이터

```python
naju2_T = naju2.T
naju2_T.drop(index='데이터기준일자', inplace=True)
naju2_T.columns = ['2021']
naju2_T.loc['주택수'] = naju2_T.loc['다세대주택'] + naju2_T.loc['단독주택'] + naju2_T.loc['다가구주택'] + naju2_T.loc['비거주용 건물내 주택'] + naju2_T.loc['연립주택']
naju3 = naju2_T.reindex(index=['가구수','주택보급률', '주택수', '다세대주택','단독주택','다가구주택','비거주용 건물내 주택','아파트','연립주택'])
naju3.index = index
naju3
```

```python
pd.concat([naju1, naju3], axis=1, join='inner')
```
